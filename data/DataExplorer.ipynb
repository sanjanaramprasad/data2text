```python
table_rows = ['sample_size', 'population', 'interventions', 'outcomes', 'punchline_text', 'punchline_effect', 
             ]

def preprocess(text):
    #text = re.sub(r'[^\w\s]','',text)
    text = text.lower()
    return text

def get_rr_data(reviews):
    src_summaries = []
    for review in reviews:
        review_summ = []
        review_summ.append(preprocess(review['sample_size']) )
        populaton_data = ' '.join([preprocess(each) for each in review['population']])
        review_summ.append(populaton_data)
        interventions = ' '.join([preprocess(each) for each in review['interventions']])
        review_summ.append(interventions)
        outcomes = ' '.join([preprocess(each) for each in review['outcomes']])
        review_summ.append(outcomes)
        review_summ.append(preprocess(review['punchline_text']))
        review_summ.append(preprocess(review['punchline_effect']))
        review_summ.append(preprocess(review['random_sequence_generation']['judgement']))
        review_summ.append(preprocess(review['allocation_concealment']['judgement']))
        review_summ.append(preprocess(review['blinding_participants_personnel']['judgement']))
        #print(review_summ)
        src_summaries.append('|'.join(review_summ))
    return src_summaries
            
def make_tabular_data(datapoints, abstract_summ):
    
    all_data_src = []
    all_data_tgt = []
    for k, v in abstract_summs['ReviewID'].items():
        rid = abstract_summs['ReviewID'][k]
        tgt_summary = abstract_summs['SummaryConclusions'][k]
        tgt_summary = preprocess(tgt_summary)
        rr_reviews = datapoints[rid]
        src_summaries = get_rr_data(rr_reviews)
        tgt_summaries = [tgt_summary] * len(src_summaries)
        all_data_src += src_summaries 
        all_data_tgt += tgt_summaries
    return all_data_src, all_data_tgt

def write_data(src_data, tgt_data, src_file, tgt_file):
    
    
    with open(src_file, 'w') as fp1:
        fp1.write('\n'.join(src_data))
    
        
    with open(tgt_file, 'w') as fp2:
        fp2.write('\n'.join(tgt_data))
         
            
```
